## Little more complex

I want max of win by runs for each season
```
For each season!
  filter season get max
```

````python
for season in df.season.unique():
    print(df[df.season == season].win_by_runs.max(), season)
````

## Well never do it!

Golden rule of thumb: If you are using for loop with pandas you are definetly doing it wrong

## How to do it?

`group_by` operation

````python
df.groupby('season').win_by_runs.max()
````

````python
df.groupby('season').win_by_runs.describe()
````

It is doing what your for loop was doing but more effiently

## Multiple groupby

What if I want max number of runs per season per toss_decsion?

````python
(
    df
    .groupby(['toss_decision', 'season'])
    .win_by_runs
    .max()
    # .reset_index()
)
````

````python
(
    df
    [["win_by_runs", "win_by_wickets", 'toss_decision', 'season']]
    .groupby(['toss_decision', 'season'])
    .agg(['min', 'max'])
    # .reset_index()
)
````

## Putting all pieces together :)

What are matches won by teams per season look like

````python
small_list = ['Sunrisers Hyderabad', 'Chennai Super Kings', 'Deccan Chargers']
win_summary_df = (
    df.groupby(["winner", "season"])
    .id
    .count()
    .reset_index()
    .rename(columns = {'id':'matches_won'})
    .query("winner in @small_list")
)
fig = px.line(win_summary_df, x="season", y="matches_won", color="winner", title='Matches Won', text='season')
fig.show()
````
